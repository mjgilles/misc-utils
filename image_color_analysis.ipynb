```python
from imutils import build_montages
from imutils import paths
import numpy as np
import argparse
import imutils
import cv2
import io
import os
```

```python
test=[]
for filename in os.listdir(os.path.join('images')):
    if filename.endswith(".jpg") or filename.endswith(".png"):
        if os.path.join(filename) not in test:
            test.append(os.path.join(filename))
```

```python
test=['Desktop/faces2/'+x for x in test]
```

```python
def image_colorfulness(image):
	# split the image into its respective RGB components
	(B, G, R) = cv2.split(image.astype("float"))
 
	# compute rg = R - G
	rg = np.absolute(R - G)
 
	# compute yb = 0.5 * (R + G) - B
	yb = np.absolute(0.5 * (R + G) - B)
 
	# compute the mean and standard deviation of both `rg` and `yb`
	(rbMean, rbStd) = (np.mean(rg), np.std(rg))
	(ybMean, ybStd) = (np.mean(yb), np.std(yb))
 
	# combine the mean and standard deviations
	stdRoot = np.sqrt((rbStd ** 2) + (ybStd ** 2))
	meanRoot = np.sqrt((rbMean ** 2) + (ybMean ** 2))
 
	# derive the "colorfulness" metric and return it
	return stdRoot + (0.3 * meanRoot)
```

```python
# initialize the results list
print("[INFO] computing colorfulness metric for dataset...")
results = []
one_C=[]
petitions=[]
# loop over the image paths
for imagePath in test:
    # load the image, resize it (to speed up computation), and
    # compute the colorfulness metric for the image

    image = cv2.imread(imagePath)
    if image is None:
        continue
    image = imutils.resize(image, width=250)
    C = image_colorfulness(image)
    one_C.append(C)
    # display the colorfulness score on the image
    cv2.putText(image, "{:.2f}".format(C), (40, 40), 
        cv2.FONT_HERSHEY_SIMPLEX, 1.4, (0, 255, 0), 3)
 
    # add the image and colorfulness metric to the results list
    results.append((image, C))
    petitions.append(imagePath.split('faces2/')[1].split('.jp')[0])
```
